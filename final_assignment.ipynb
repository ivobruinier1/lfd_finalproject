```python
  # -*- coding: utf-8 -*-
"""lfd_assignment3_PLMS.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1XuRTzJzJnePwtfDHIQ0eRZg6BN8r0k3y
"""

import argparse
import random
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
import tensorflow as tf
from transformers import TFAutoModelForSequenceClassification, AutoTokenizer
from tensorflow.keras.losses import CategoricalCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler

# Make reproducible
np.random.seed(1234)
tf.random.set_seed(1234)
random.seed(1234)

def read_corpus(corpus_file):
    """Reads the corpus file and splits lines into lists of
    documents and corresponding lists of labels. If use_sentiment
    is true, use the binary (sentiment) labels. Else, use the multi-
    class (category) labels."""
    documents, labels = [], []
    with open(corpus_file, encoding='utf-8') as f:
        for line in f:
            tokens = line.strip().split('\t')
            documents.append(tokens[0])
            labels.append(tokens[1])
    return documents, labels

# Function to tokenize data
def tokenize_data(tokenizer, data, max_length):
    return tokenizer(data, padding=True, max_length=max_length, truncation=True, return_tensors="np").data

# Define a learning rate schedule function
def lr_schedule(epoch, initial_lr=5e-5, drop=0.5, epochs_drop=5):
    return initial_lr * (drop ** (epoch // epochs_drop))

def experiment_with_model(model_name, max_length, batch_size, epochs):
    # Load the model and tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = TFAutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

    # Read the data
    X_train, Y_train = read_corpus('train.tsv')
    X_dev, Y_dev = read_corpus('dev.tsv')
    X_test, Y_test = read_corpus('test.tsv')

    # Transform string labels to one-hot encodings
    encoder = LabelBinarizer()
    Y_train_bin = encoder.fit_transform(Y_train).flatten()
    Y_dev_bin = encoder.transform(Y_dev).flatten()
    Y_test_bin = encoder.transform(Y_test).flatten()

    # Tokenize the data
    tokens_train = tokenize_data(tokenizer, X_train, max_length)
    tokens_dev = tokenize_data(tokenizer, X_dev, max_length)
    tokens_test = tokenize_data(tokenizer, X_test, max_length)

    # Compile the model
    loss_function = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
    optim = Adam(learning_rate=5e-5)
    model.compile(loss=loss_function, optimizer=optim, metrics=['accuracy'])

    # Learning rate scheduler (drops learning rate after a given number of epochs)
    lr_scheduler = LearningRateScheduler(lambda epoch: lr_schedule(epoch))

    # Train the model
    history = model.fit(tokens_train, Y_train_bin, validation_data=(tokens_dev, Y_dev_bin),
                        epochs=epochs, batch_size=batch_size, callbacks=[EarlyStopping(patience=5), lr_scheduler]) # here you can change the patience of the learning rate scheduler

    # Evaluate the model on the dev set
    Y_pred = model.predict(tokens_dev)["logits"]
    Y_pred = np.argmax(Y_pred, axis=1)
    dev_accuracy = accuracy_score(Y_dev_bin, Y_pred)

    print(f'Accuracy on validation set with model {model_name}: {dev_accuracy:.3f}')

    # Evaluate the model on the test set
    Y_pred_test = model.predict(tokens_test)["logits"]
    Y_pred_test = np.argmax(Y_pred_test, axis=1)  # Get predicted classes (indices)
    test_accuracy = accuracy_score(Y_test_bin, Y_pred_test)

    print(f'Accuracy on test set with model {model_name}: {test_accuracy:.3f}')

    # Classification report for test set
    print('\n*** CLASSIFICATION REPORT ***')
    print(classification_report(Y_test_bin, Y_pred_test))

    # Confusion matrix for test set
    print('\n*** CONFUSION MATRIX ***')

    # Use the original string labels (not one-hot encoded)
    labels = encoder.classes_  # ['books', 'camera', 'dvd', 'health', 'music', 'software']

    # Calculates and print the confusion matrix
    cm = confusion_matrix(Y_test_bin, Y_pred_test, labels=np.arange(len(labels)))

    # Prints confusion matrix in readable format
    print(' '.join(labels))
    print(cm)

    # This will plot the confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()

    return history, test_accuracy



def main():
    # Example experiment
    model_names = ["bert-base-uncased", "google/electra-base-discriminator", "GroNLP/bert-base-dutch-cased", "dbmdz/bert-base-german-cased", "dccuchile/bert-base-spanish-wwm-cased", "camembert-base" ]  # # Here you can add more models through the huggingface library
    results = {}

    for model_name in model_names:
        print(f"Experimenting with model: {model_name}")
        history, test_accuracy = experiment_with_model(model_name=model_name, max_length=128, batch_size=32, epochs=2) # Here you can fine tune various settings like epochs, batch size, etc.
        results[model_name] = {'history': history, 'test_accuracy': test_accuracy}


if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python

'''
LfD Assignment 3 - LSTM and Pretrained models

This script allows the user to run a LSTM classifier or Pretrained Language Model.

Available command-line options:
-i  Input file to learn from (default train.txt)
-d  Separate dev set to read in (default dev.txt)
-t  If added, use trained model to predict on test set
-e  Embedding file we are using (default glove_reviews.json)

--no_pre_embeddings         Do not use pretrained embeddings (as is default).
--trainable                 Set Trainable to True in the embedding layer. Default is False.
--extra_dense               Add an extra Dense layer between the Embedding and first LSTM layer.
--more_layers               Add an extra LSTM layer that improves performance
--dropout                   Use dropout. No dropout is used by default.
--recurrent_dropout         Use recurrent dropout. No dropout is used by default.
--optimizer                 The optimizer to use. Options: SGD, RMSprop, Adam, AdamW
--bidirectional             Change default LSTM to a bidirectional model.
--learning_rate             Set the learning rate. Default: 0.01
--epochs                    Set the number of epochs. Default: 50
--batch_size                Set the batch size. Default: 8

How to run best model:
python lfd_assignment3.py --optimizer RMSprop --learning_rate 0.005 --batch_size 32 --dropout 0.5 --bidirectional
python3 lfd_assignment3.py --optimizer RMSprop --learning_rate 0.005 --batch_size 32 --dropout 0.5 --bidirectional --trainable -t test.txt
'''

import random as python_random
import json
import argparse
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Embedding, LSTM, Bidirectional, Input
from keras.initializers import Constant
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelBinarizer
from tensorflow.keras.optimizers import SGD, RMSprop, Adam, AdamW
from tensorflow.keras.layers import TextVectorization
import tensorflow as tf
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt


def create_arg_parser():
    parser = argparse.ArgumentParser()
    parser.add_argument("-i", "--train_file", default='train.tsv', type=str,
                        help="Input file to learn from (default train.tsv)")
    parser.add_argument("-d", "--dev_file", type=str, default='dev.tsv',
                        help="Separate dev set to read in (default dev.tsv)")
    parser.add_argument("-t", "--test_file", type=str, default='test.tsv',
                        help="If added, use trained model to predict on test set")
    parser.add_argument("-e", "--embeddings", default='glove.6B.100d.json', type=str,
                        help="Embedding file we are using (default glove.6B.100d.jsonn)")
    parser.add_argument("--no_pre_embeddings", action="store_true",
                        help="Do not use pre-trained embeddings (as is default)")
    parser.add_argument("--trainable", action="store_true",
                        help="Set Trainable to True in the embedding layer. Default is False.")
    parser.add_argument("--extra_dense", action="store_true",
                        help="Add an extra Dense layer between the Embedding and first LSTM layer.")
    parser.add_argument("--more_layers", action="store_true",
                        help="Add an extra LSTM layer that improves performance")
    parser.add_argument("--dropout", default="0", type=float,
                        help="Add dropout (float between 0-1). Default: 0")
    parser.add_argument("--recurrent_dropout", default="0", type=float,
                        help="Add recurrent dropout (float between 0-1). Default: 0")
    parser.add_argument("--optimizer", type=str, default="SGD", choices=['SGD', 'RMSprop', 'Adam', 'AdamW'],
                        help="The optimizer to use. SGD is default. Options: SGD, RMSprop, Adam, AdamW")
    parser.add_argument("--bidirectional", action="store_true", help="Change LSTM to a bidirectional model.")
    parser.add_argument("--learning_rate", type=float, default=0.01, help="Change learning rate. 0.01 default.")
    parser.add_argument("--epochs", type=int, default=1, help="Change number of epochs. 50 default.")
    parser.add_argument("--batch_size", type=int, default=8, help="Change number of epochs. 8 default.")
    args = parser.parse_args()
    return args


def read_corpus(corpus_file):
    '''Read in review data set and returns docs and labels'''
    documents, labels = [], []
    with open(corpus_file, encoding='utf-8') as f:
        for line in f:
            tokens = line.strip().split('\t')
            documents.append(tokens[0])
            labels.append(tokens[1])
    return documents, labels


def read_embeddings(embeddings_file):
    '''Read in word embeddings from file and save as numpy array'''
    embeddings = json.load(open(embeddings_file, 'r'))
    return {word: np.array(embeddings[word]) for word in embeddings}


def get_emb_matrix(voc, emb):
    '''Get embedding matrix given vocab and the embeddings'''
    num_tokens = len(voc) + 2
    word_index = dict(zip(voc, range(len(voc))))
    # Bit hacky, get embedding dimension from the word "the"
    embedding_dim = len(emb["the"])
    # Prepare embedding matrix to the correct size
    embedding_matrix = np.zeros((num_tokens, embedding_dim))
    for word, i in word_index.items():
        embedding_vector = emb.get(word)
        if embedding_vector is not None:
            # Words not found in embedding index will be all-zeros.
            embedding_matrix[i] = embedding_vector
    # Final matrix with pretrained embeddings that we can feed to embedding layer
    return embedding_matrix


def create_model(Y_train, emb_matrix, args):
    '''Create the Keras model to use'''
    # Define settings, you might want to create cmd line args for them
    learning_rate = args.learning_rate
    loss_function = 'binary_crossentropy'

    # Pick the correct optimizer
    if args.optimizer == 'SGD':
        optim = SGD(learning_rate=learning_rate)
    elif args.optimizer == 'RMSprop':
        optim = RMSprop(learning_rate=learning_rate)
    elif args.optimizer == 'Adam':
        optim = Adam(learning_rate=learning_rate)
    elif args.optimizer == 'AdamW':
        optim = AdamW(learning_rate=learning_rate)

    # Take embedding dim and size from emb_matrix
    embedding_dim = len(emb_matrix[0])
    num_tokens = len(emb_matrix)
    num_labels = len(set(Y_train))

    # Now build the model
    model = Sequential()
    # If pre_trained embeddings should be used (default), add them.
    if args.no_pre_embeddings:
        model.add(Embedding(num_tokens, embedding_dim, trainable=args.trainable))
    else:
        model.add(Embedding(num_tokens, embedding_dim, embeddings_initializer=Constant(emb_matrix),
                            trainable=args.trainable))

    # If parameter extra_dense is set, add an extra dense layer
    if args.extra_dense:
        print('Adding extra dense layer')
        model.add(Dense(input_dim=embedding_dim, units=num_labels, activation="sigmoid"))

    # Add more LSTM layers if commandline option is given
    if args.more_layers:
        if args.bidirectional:
            model.add(Bidirectional(LSTM(units=num_labels, return_sequences=True, dropout=args.dropout,
                                         recurrent_dropout=args.recurrent_dropout, activation="elu")))
            model.add(Bidirectional(LSTM(units=num_labels, dropout=args.dropout,
                                         recurrent_dropout=args.recurrent_dropout, activation="elu")))
        else:
            model.add(LSTM(units=num_labels, return_sequences=True, dropout=args.dropout,
                           recurrent_dropout=args.recurrent_dropout, activation="elu"))
            model.add(LSTM(units=num_labels, dropout=args.dropout,
                           recurrent_dropout=args.recurrent_dropout, activation="elu"))
    else:
        # Change LSTM to a bidirectional model
        if args.bidirectional:
            model.add(Bidirectional(LSTM(units=num_labels, dropout=args.dropout,
                                         recurrent_dropout=args.recurrent_dropout, activation="elu")))
        else:
            model.add(LSTM(units=num_labels, dropout=args.dropout,
                           recurrent_dropout=args.recurrent_dropout, activation="elu"))

    # Ultimately, end with dense layer with softmax
    model.add(Dense(units=1, activation='sigmoid'))
    # Compile model using our settings, check for accuracy
    model.compile(loss=loss_function, optimizer=optim, metrics=['accuracy'])
    return model


def train_model(model, X_train, Y_train, X_dev, Y_dev, args, encoder):
    '''Train the model here. Note the different settings you can experiment with!'''
    # Potentially change these to cmd line args again
    # And yes, don't be afraid to experiment!
    verbose = 1
    batch_size = args.batch_size
    epochs = args.epochs

    # Early stopping: stop training when there are three consecutive epochs without improving
    # It's also possible to monitor the training loss with monitor="loss"
    callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
    # Finally fit the model to our data
    model.fit(X_train, Y_train, verbose=verbose, epochs=epochs, callbacks=[callback], batch_size=batch_size, validation_data=(X_dev, Y_dev))
    # Print final accuracy for the model (clearer overview)
    test_set_predict(model, X_dev, Y_dev, "dev", args, encoder)
    return model


def test_set_predict(model, X_test, Y_test, ident, args, encoder):
    '''Do predictions and measure accuracy on our own test set (that we split off train)'''
    # Get predictions using the trained model
    Y_pred = model.predict(X_test)
    # Finally, convert to numerical labels to get scores with sklearn
    Y_pred = (Y_pred > 0.5).astype("int32")

    # If you have gold data, you can calculate accuracy
    Y_test = (Y_test > 0.5).astype("int32")

    print('Accuracy on own {1} set: {0}'.format(round(accuracy_score(Y_test, Y_pred), 3), ident))

    # Print detailed results
    print_evaluation(Y_test, Y_pred, encoder)


def print_evaluation(Y_test, Y_pred, encoder):
    """Takes true labels and predicted labels and
    prints evaluation measures (a classification report
    and confusion matrix)"""
    labels = encoder.classes_

    print('\n*** CLASSIFICATION REPORT ***')
    print(classification_report(Y_test, Y_pred))

    print('\n*** CONFUSION MATRIX ***')
    cm = confusion_matrix(Y_test, Y_pred)
    print(' '.join(labels))
    print(cm)

    # Pretty confusion matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
    disp.plot(cmap=plt.cm.Blues)
    plt.show()


def main():
    '''Main function to train and test neural network given cmd line arguments'''
    # Make reproducible as much as possible
    tf.keras.utils.set_random_seed(1234)
    python_random.seed(1234)

  # Instead of using argparse, create an object with the desired arguments
    class Args:
        train_file = 'train.tsv'
        dev_file = 'dev.tsv'
        test_file = 'test.tsv'
        embeddings = 'glove.6B.100d.json'
        no_pre_embeddings = False  # Set to True if you don't want to use pre-trained embeddings
        trainable = False         # Set to True to make embeddings trainable
        extra_dense = False      # Set to True to add an extra dense layer
        more_layers = False       # Set to True to add more LSTM layers
        dropout = 0.5             # Dropout rate
        recurrent_dropout = 0       # Recurrent dropout rate
        optimizer = 'Adam'      # Optimizer to use
        bidirectional = True      # Use bidirectional LSTM
        learning_rate = 0.005     # Learning rate
        epochs = 50                # Number of epochs
        batch_size = 32           # Batch size

        def print_settings(cls):
          print("Training Settings:")
          print("------------------")
          # Iterate through class attributes and print their names and values
          for attr, value in cls.__dict__.items():
              if not attr.startswith('__') and not callable(value):  # Filter out special attributes and methods
                  print(f"{attr}: {value}")



    args = Args()  # Use the Args object instead of argparse.ArgumentParser().parse_args()
    Args.print_settings(Args)

    # Read in the data and embeddings
    X_train, Y_train = read_corpus(args.train_file)
    X_dev, Y_dev = read_corpus(args.dev_file)
    X_test, Y_test = read_corpus(args.test_file)
    embeddings = read_embeddings(args.embeddings)

    # Transform words to indices using a vectorizer
    vectorizer = TextVectorization(standardize=None, output_sequence_length=50)
    # Use train and dev to create vocab - could also do just train
    text_ds = tf.data.Dataset.from_tensor_slices(X_train + X_dev)
    vectorizer.adapt(text_ds)
    # Dictionary mapping words to idx
    voc = vectorizer.get_vocabulary()
    emb_matrix = get_emb_matrix(voc, embeddings)

     # Transform string labels to one-hot encodings
    encoder = LabelBinarizer()
    Y_train_bin = encoder.fit_transform(Y_train)
    Y_dev_bin = encoder.fit_transform(Y_dev)
    Y_test_bin = encoder.fit_transform(Y_test)

    # Create model
    model = create_model(Y_train, emb_matrix, args)


    # Transform input to vectorized input
    X_train_vect = vectorizer(np.array([[s] for s in X_train])).numpy()
    X_dev_vect = vectorizer(np.array([[s] for s in X_dev])).numpy()
    X_test_vect = vectorizer(np.array([[s] for s in X_test])).numpy()

    # Train the model
    model = train_model(model, X_train_vect, Y_train_bin, X_dev_vect, Y_dev_bin, args, encoder)

    test_set_predict(model, X_test_vect, Y_test_bin, "test", args, encoder)


main()
```
